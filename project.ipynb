##### Oppgave 1a:

Vi vil vise at 
\begin{align*}
    \phi (x,t) = \frac{1}{\sqrt{4 \pi Dt}} \int_{-\infty}^{\infty} h(y) e^{-\frac{{x-y}^2}{4Dt}} dy
\end{align*}

er en løsning av diffusjonsligningen

\begin{align*}
    \frac{\partial \phi}{\partial t} = D \frac{{\partial}^2 \phi}{\partial x^2},
\end{align*}

der D = $\frac{d h^2}{2 \Delta t} > 0$


Vi definerer

\begin{align*}
G(\xi, t)= \frac{1}{\sqrt{4 \pi Dt}} e^{-\frac{\xi ^2}{4Dt}},
\end{align*}

som er en Gauss-fordeling. Her beskrives hvordan sannsynligheten for å finne en partikkel som sprer seg over tid. Setter x-y til $\xi$ for å gjøre det lettere siden disse ikke har noe påvirkning på derivasjon. 

Da kan løsningen skrives som

\begin{align*}
    \phi (x,t) = \int_{-\infty}^{\infty} h(y) G(\xi, t) dy.
\end{align*}


Dette betyr at løsningen består av bidrag fra alle startposisjoner y, der hvert bidrag er vektet med sannsynligheten for at en partikkel som starter i y befinner seg i x etter tiden t.

Det kan vises ved direkte derivasjon at

\begin{align*}
    \frac{\partial G}{\partial t} = D \frac{\partial ^2 G}{\partial \xi ^2}
\end{align*}


Videre skriver vi 

\begin{align*}
    G(\xi, t) = u(t) e^{-v(t) \xi^2},
\end{align*}

der

\begin{align*}
    u(t) = (4\pi Dt)^{-\frac{1}{2}}, v(t)=\frac{1}{4Dt}.
\end{align*}

dette for å gjøre utregningen lettere.

Vi deriverer først med hensyn på $\xi$:

\begin{align*}
    \frac{\partial G}{\partial \xi} = u(t) e^{-v(t) \xi^2} (-2v(t)\xi)
\end{align*}

Siden $2v(t) = \frac{1}{2Dt}$, får vi 

\begin{align*}
    \frac{\partial G}{\partial \xi}= -\frac{\xi}{2Dt} G.
\end{align*}

Vi deriverer en gang til:

\begin{align*}
    \frac{\partial^2 G}{\partial \xi^2}=\frac{\partial}{\partial \xi} (-\frac{\xi}{2Dt} G)
\end{align*}

Ved bruk av produktregelen får vi

\begin{align*}
    \frac{\partial^2 G}{\partial \xi^2} &= (-\frac{\xi}{2Dt}) \cdot (u(t)e^{-v(t)\xi^2}) \\
    &=-\frac{1}{2Dt} (G) + (-\frac{\xi}{4Dt})(-\frac{\xi}{4Dt} G) \\
    &= -\frac{1}{2Dt} G + \frac{\xi^2}{4D^2 t^2} G.
\end{align*}


Dette kan skrives som

\begin{align*}
    \frac{\partial^2 G}{\partial \xi^2} = G(-\frac{1}{2Dt}+\frac{\xi^2}{4D^2 t^2})
\end{align*}

Ved å multiplisere med D får vi

\begin{align*}
    D \frac{\partial^2 G}{\partial \xi^2} = G(-\frac{1}{2t} + \frac{\xi^2}{4Dt^2})
\end{align*}

da har vi den ene siden. Og nå kan vi finne den andre siden. Så vi deriverer med hensyn på t:

\begin{align*}
    \frac{\partial G(\xi, t)}{\partial t} = \frac{\partial}{\partial t} \frac{1}{\sqrt{4\pi Dt}} e^{-\frac{\varepsilon^2}{4Dt}}:
\end{align*}

Vi definerer

\begin{align*}
    u(t) &= (4\pi Dt)^{\frac{1}{2}}, \\
    v(t) &= e^{-\frac{\varepsilon^2}{4Dt}}
\end{align*}

Derivasjon av u(t) gir

\begin{align*}
    u'(t)=-\frac{1}{2} (4\pi D)^{-\frac{1}{2}} (t)^{-\frac{3}{2}} = -\frac{1}{2t} (4\pi Dt)^{-\frac{1}{2}}.
\end{align*}

Derivasjon av v(t), her bruker vi kjerneregelen, der $v(t)=e^{u(t)}$.

Da er
\begin{align*}
    v'(t)=e^{u(t)}\cdot u'(t)
\end{align*}

Videre har vi

\begin{align*}
    g(u)= e^u , g'(u)=e^u .
\end{align*}

Dermed får vi

\begin{align*}
    v'(t)=e^{u(t)} \cdot \frac{\varepsilon^2}{4Dt^2}.
\end{align*}

Setter vi inn uttrykket for u(t), får vi

\begin{align*}
    v'(t)=e^{-\frac{\varepsilon^2}{4Dt}} (\frac{\varepsilon^2}{4Dt^2})
\end{align*}

Så setter vi det inn i produktformelene. Videre har vi at 

\begin{align*}
    u'v + uv'
\end{align*}

\begin{align*}
    = -\frac{1}{2t} (4\pi Dt)^{-\frac{1}{2}} \cdot (e^{-\frac{\varepsilon^2}{4Dt}}) + (4\pi Dt)^{-\frac{1}{2}} (e^{-\frac{\varepsilon^2}{4Dt}}(\frac{\varepsilon^2}{4Dt^2})) \\
    =(4\pi Dt)^{-\frac{1}{2}} e^{-\frac{\varepsilon^2}{4Dt}} (-\frac{1}{2t} + \frac{\varepsilon^2}{4Dt^2}) \\
    \frac{\partial G}{\partial t} = G(\varepsilon, t)(-\frac{1}{2t} + \frac{\varepsilon^2}{4Dt^2})
\end{align*}

Er Identisk så oppfyller G diffusjonsligningen. 

Viser så at $\phi$ løser diffusjonsligningen:

\begin{align*}
    \phi(x,t) &= \int_{-\infty}^{infty} h(y) G(x-y,t)dy \\
    \frac{\partial^2 \phi}{\partial x^2} &= \int h(y) \frac{\partial^2}{\partial x^2} G(x-y,t)dy
\end{align*}

Når vi deriverer $G(x-y,t)$ er det samme som å derivere $G(\xi, t)$, $(\xi = x - y)$

\begin{align*}
    \text{Altså:} \frac{\partial^2}{\partial x^2} G(x-y,t)=\frac{\partial^2}{\partial \xi^2} G(\xi, t) \\
    \text{Vi vet G oppfyller }\partial_t G = D \partial_{\xi \xi} G \text{ får vi: } \\
    \frac{\partial \phi}{\partial t} = \int h(y) D\frac{\partial^2}{\partial x^2} G(x-y,t)dy=D\frac{\partial^2 \phi}{\partial x^2} \\
    \Rightarrow \text{Dermed er }\phi \text{ en løsning av diffusjonsligningen.}
\end{align*}


Fysisk tolkning av h(y): Når $t \rightarrow 0^+$ blir $G(x-y,t)$ en veldig smal normalfordeling rundt $x=y$ (Dirachs deltafunksjon). Dermed gjelder

\begin{align*}
    G(\cdot , t) \rightarrow \delta(\cdot).
\end{align*}

Da får vi

\begin{align*}
    \phi (x,t) = \int_{-\infty}^{\infty} h(y) G(x-y, t)dy \xrightarrow[t \to 0^{+}] \int_{-\infty}^{infty} h(y) \delta (x-y) dy = h(x).
\end{align*}

Her er h(x) initialfordelingen av partikler ved starttid, det vil si

\begin{align*}
    \phi (x,0) = h(x)
\end{align*}

Altså beskriver h(y) hvor sannsynlig det er å finne en partikkel ved posisjon y når t = 0.

##### Oppgave 1b:

En partikkel starter på $x_0$ og med et steg, så kan den ende opp på $x$ $\epsilon$ \{$x_0 - 1$, $x_0$ , $x_0 + 1$\}.


Anta at for $P: x_0 \to x$, så er sannsynligheten for å finne en partikkel x i V(x) $P(x)=\frac{e^{-\beta V(x)}}{Z}$.



Vi antar også at det finnes en proporsjonal constant som er lik for alle steg muligheter.

La

\begin{align*}
    p^- =CP(x_0 -1), p^0 =CP(x_0), og p^+ =CP(x_0 +1)
\end{align*}

Her er $p^- + p^0 + p^+ =1$, og dette kan brukes for å løse for C. 

\begin{align*}
    1=C[P(x_0-1)+P(x_0)+P(x_0+1)]
\end{align*}

\begin{align*}
    C=\frac{1}{P(x_0-1)+P(x_0)+P(x_0+1)}
\end{align*}


Nå setter vi inn $\{x_0-1, x_0, x_0+1\}$ i P(x).

\begin{align*}
    P(x_0-1)=\frac{e^{-\beta V(x_0-1)}}{Z}  \\
    P(x_0)=\frac{e^{-\beta V(x_0)}}{Z}  \\
    P(x_0+1)=\frac{e^{-\beta V(x_0+1)}}{Z}
\end{align*}

Alle har felles nevner, så,

\begin{equation*}
    C=\frac{Z}{e^{-\beta V(x_0-1)}+e^{-\beta V(x_0)}+e^{-\beta V(x_0+1)}}
\end{equation*}


Nå kan vi sette konstanten vår C inn i $\{p^-,p^0,p^+\}$:

$p^-$:

\begin{align*}
    p^-=CP(x_0-1)=\frac{Z}{\Sigma e^{-\beta V}}\cdot\frac{e^{-\beta V(x_0-1)}}{Z}
\end{align*}

Z kansellerer, så vi får at

\begin{equation*}
    p^-=\frac{e^{-\beta V(x_0-1)}}{e^{-\beta V(x_0-1)}+e^{-\beta V(x_0)}+e^{-\beta V(x_0+1)}}
\end{equation*}

Deler både telleren og nevneren med $e^{-\beta V(x_0-1)}$


\begin{align*}
    \frac{e^{-\beta V(x_0-1)}}{e^{-\beta V(x_0-1)}}&=1   \\
    \frac{e^{-\beta V(x_0)}}{e^{-\beta V(x_0-1)}}&=e^{-\beta [V(x_0)-V(x_0-1)]}  \\
    \frac{e^{-\beta V(x_0+1)}}{e^{-\beta V(x_0-1)}}&=e^{-\beta [V(x_0+1)-V(x_0-1)]}
\end{align*}


Da får vi at 

\begin{equation*}
    p^-=\frac{1}{1+e^{-\beta [V(x_0)-V(x_0-1)]}+e^{-\beta [V(x_0+1)-V(x_0-1)]}}
\end{equation*}

Ved å gjøre dette med $p^0$ og $p^+$ så får vi at 

\begin{equation*}
    p^0=\frac{1}{1+e^{-\beta [V(x_0-1)-V(x_0)]}+e^{-\beta [V(x_0+1)-V(x_0)]}}
\end{equation*}

og

\begin{equation*}
    p^+=\frac{1}{1+e^{-\beta [V(x_0-1)-V(x_0+1)]}+e^{-\beta [V(x_0)-V(x_0+1)]}}
\end{equation*}


Dette var det vi skulle vise. $\square$



##### Oppgave 1c:

Skal finne ut av hva som skjer med sannsynlighetene $\{p^+,p^0,p^-\}$ fra oppgave 1b når temperaturen T oppfyller de følgende betingelsene: 

\begin{align*}
    \bullet \qquad& k_BT>>|V(x+1)-V(x)| \text{ for noen x}, \\
    \bullet \qquad& k_BT<<|V(x+1)-V(x) \text{ for noen x}.
\end{align*}

Boltzmanns vekt/faktor er definert som $e^{-\beta E_i}=e^{\frac{-E_i}{k_BT}}$. Dette ligner på sannsynligheten vi fikk i ligning 7 i oppgave 1b. 

Så, vi kan skrive om resultatene fra 1B om til Boltzmanns vekt form. 

La $w^+=e^{-\beta V(x_0+1)}$, $w^0=e^{-\beta V(x_0)}$, $w^-=e^{-\beta V(x_0-1)}$. 

Vi får da at $p^+=\frac{w^+}{w^-+w^0+w^+}$, $p^0=\frac{w^0}{w^-+w^0+w^+}$, og $p^-=\frac{w^-}{w^-+w^0+w^+}$.


##### $k_BT>>|V(x+1)-V(x)|$ for noen x:

Hvis T oppfyller $k_BT>>|V(x+1)-V(x)|$, så betyr det at vi ser på en høyere temperatur limit. 

Når temperaturen er høy, så er energiforskjellene mellom nabo punkter små når de er sammenlignet mde den termiske energien. Da er

\begin{equation*}
    e^{-\beta [V(x^i)-V(x^j)]}\approx 1
\end{equation*} 

for alle naboposisjoner. 

Her er, $w^- \approx w^0 \approx w^+$, så vi får at

\begin{align*}
    p^+ \approx p^0 \approx p^- \approx \frac{1}{3}
\end{align*}

Her kan vi se at partikkelen utfører tilnærmet en symmetrisk tilfeldig vandring, så potensialet har liten betydning for bevegelsen. 


##### $k_BT<<|V(x+1)-V(x)|$ for noen x

For lave temperaturer, så blir $\beta$ stor. Dette betyr at små energiforskjeller gir store forskjeller i Boltzmann faktorene. Da vil posisjonen med lavest potensialenergi dominere normaliseringen. 

Sannsynligheten for å bevege seg til posisjon med lavest $V(x)$ blant $\{x_0-1,x_0,x_0+1\}$ går mot 1, mens de andre går mot 0. 

Dette betyr at bevegelsen blir tilnærmet deterministisk, og partikkelen beveger seg nedover i potensiallandskapet. 


Så, i konklusjon så ser vi at

\begin{align*}
    \bullet \qquad &\text{ For høy temperatur, så kan potensialet neglisjeres, og problemet forenkles til vanlig diffusjon} \\
    &\text{ med konstante overgangssansynligheter.} \\
    \bullet \qquad &\text{ For lav temperatur, så kan den termiske støyen neglisjeres, og bevegelsen kan forenkles til at} \\
    &\text{ partikkelen alltid beveger seg mot lavere potensialenergi. }
\end{align*}

```python
### Exercise 2a)

N_particles = 101
N_steps = 200
h = 1
beta_k = 1
N_runs = 100

initial_positions = np.arange(N_particles)

#skal starte med i) V(x)=k:

def V_constant(x,k=1):
    return k

def transition_probabilities(x0, V, beta=1.0):
    V_minus = V(x0 - 1)
    V_0 = V(x0)
    V_plus = V(x0 + 1)

    w_minus = np.exp(-beta * V_minus)
    w_0 = np.exp(-beta * V_0)
    w_plus = np.exp(-beta * V_plus)

    Z = w_minus + w_0 + w_plus

    p_minus = w_minus / Z
    p_0 = w_0 / Z
    p_plus = w_plus / Z

    return p_minus, p_0, p_plus

def step(positions, V, beta=1.0):
    new_positions = positions.copy()
    for i, x0 in enumerate(positions):
        p_minus, p_0, p_plus = transition_probabilities(x0, V, beta)
        r = np.random.rand()
        if r <= p_minus:
            new_positions[i] = x0 - h
        elif r > 1 - p_plus:
            new_positions[i] = x0 + h
        else:
            new_positions[i] = x0
    return new_positions

counts_sum = {}
min_x = None
max_x = None

for _ in range(N_runs):
    positions = initial_positions.copy()

    for _ in range(N_steps):
        positions = step(positions, V_constant, beta=beta_k)

    xs, counts = np.unique(positions, return_counts=True)

    if min_x is None:
        min_x, max_x = int(xs.min()), int(xs.max())
    else:
        min_x = min(min_x, int(xs.min()))
        max_x = max(max_x, int(xs.max()))

    for x, c in zip(xs, counts):
        counts_sum[int(x)] = counts_sum.get(int(x), 0) + int(c)
```

```python
#plot for 2a(i):

x_axis = np.arange(min_x, max_x + 1)
avg_counts = np.array([counts_sum.get(int(x), 0) for x in x_axis], dtype=float) / N_runs
avg_density = avg_counts / N_particles


plt.figure()
plt.plot(x_axis, avg_density, marker='o', linestyle='-')
plt.xlabel("x")
plt.ylabel("Gjennomsnittlig partikkelfordeling (sannsynligheten)")
plt.title("Oppgave 2a(i): $V(x) = k$")
plt.grid(True, alpha=0.3)
plt.show()
```
