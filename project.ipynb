##### Oppgave 1a:

Vi vil vise at 
\begin{align*}
    \phi (x,t) = \frac{1}{\sqrt{4 \pi Dt}} \int_{-\infty}^{\infty} h(y) e^{-\frac{{x-y}^2}{4Dt}} dy
\end{align*}

er en løsning av diffusjonsligningen

\begin{align*}
    \frac{\partial \phi}{\partial t} = D \frac{{\partial}^2 \phi}{\partial x^2},
\end{align*}

der D = $\frac{d h^2}{2 \Delta t} > 0$


Vi definerer

\begin{align*}
G(\xi, t)= \frac{1}{\sqrt{4 \pi Dt}} e^{-\frac{\xi ^2}{4Dt}},
\end{align*}

som er en Gauss-fordeling. Her beskrives hvordan sannsynligheten for å finne en partikkel som sprer seg over tid. Setter x-y til $\xi$ for å gjøre det lettere siden disse ikke har noe påvirkning på derivasjon. 

Da kan løsningen skrives som

\begin{align*}
    \phi (x,t) = \int_{-\infty}^{\infty} h(y) G(\xi, t) dy.
\end{align*}


Dette betyr at løsningen består av bidrag fra alle startposisjoner y, der hvert bidrag er vektet med sannsynligheten for at en partikkel som starter i y befinner seg i x etter tiden t.

Det kan vises ved direkte derivasjon at

\begin{align*}
    \frac{\partial G}{\partial t} = D \frac{\partial ^2 G}{\partial \xi ^2}
\end{align*}


Videre skriver vi 

\begin{align*}
    G(\xi, t) = u(t) e^{-v(t) \xi^2},
\end{align*}

der

\begin{align*}
    u(t) = (4\pi Dt)^{-\frac{1}{2}}, v(t)=\frac{1}{4Dt}.
\end{align*}

dette for å gjøre utregningen lettere.

Vi deriverer først med hensyn på $\xi$:

\begin{align*}
    \frac{\partial G}{\partial \xi} = u(t) e^{-v(t) \xi^2} (-2v(t)\xi)
\end{align*}

Siden $2v(t) = \frac{1}{2Dt}$, får vi 

\begin{align*}
    \frac{\partial G}{\partial \xi}= -\frac{\xi}{2Dt} G.
\end{align*}

Vi deriverer en gang til:

\begin{align*}
    \frac{\partial^2 G}{\partial \xi^2}=\frac{\partial}{\partial \xi} (-\frac{\xi}{2Dt} G)
\end{align*}

Ved bruk av produktregelen får vi

\begin{align*}
    \frac{\partial^2 G}{\partial \xi^2} &= (-\frac{\xi}{2Dt}) \cdot (u(t)e^{-v(t)\xi^2}) \\
    &=-\frac{1}{2Dt} (G) + (-\frac{\xi}{4Dt})(-\frac{\xi}{4Dt} G) \\
    &= -\frac{1}{2Dt} G + \frac{\xi^2}{4D^2 t^2} G.
\end{align*}


Dette kan skrives som

\begin{align*}
    \frac{\partial^2 G}{\partial \xi^2} = G(-\frac{1}{2Dt}+\frac{\xi^2}{4D^2 t^2})
\end{align*}

Ved å multiplisere med D får vi

\begin{align*}
    D \frac{\partial^2 G}{\partial \xi^2} = G(-\frac{1}{2t} + \frac{\xi^2}{4Dt^2})
\end{align*}

da har vi den ene siden. Og nå kan vi finne den andre siden. Så vi deriverer med hensyn på t:

\begin{align*}
    \frac{\partial G(\xi, t)}{\partial t} = \frac{\partial}{\partial t} \frac{1}{\sqrt{4\pi Dt}} e^{-\frac{\varepsilon^2}{4Dt}}:
\end{align*}

Vi definerer

\begin{align*}
    u(t) &= (4\pi Dt)^{\frac{1}{2}}, \\
    v(t) &= e^{-\frac{\varepsilon^2}{4Dt}}
\end{align*}

Derivasjon av u(t) gir

\begin{align*}
    u'(t)=-\frac{1}{2} (4\pi D)^{-\frac{1}{2}} (t)^{-\frac{3}{2}} = -\frac{1}{2t} (4\pi Dt)^{-\frac{1}{2}}.
\end{align*}

Derivasjon av v(t), her bruker vi kjerneregelen, der $v(t)=e^{u(t)}$.

Da er
\begin{align*}
    v'(t)=e^{u(t)}\cdot u'(t)
\end{align*}

Videre har vi

\begin{align*}
    g(u)= e^u , g'(u)=e^u .
\end{align*}

Dermed får vi

\begin{align*}
    v'(t)=e^{u(t)} \cdot \frac{\varepsilon^2}{4Dt^2}.
\end{align*}

Setter vi inn uttrykket for u(t), får vi

\begin{align*}
    v'(t)=e^{-\frac{\varepsilon^2}{4Dt}} (\frac{\varepsilon^2}{4Dt^2})
\end{align*}

Så setter vi det inn i produktformelene. Videre har vi at 

\begin{align*}
    u'v + uv'
\end{align*}

\begin{align*}
    = -\frac{1}{2t} (4\pi Dt)^{-\frac{1}{2}} \cdot (e^{-\frac{\varepsilon^2}{4Dt}}) + (4\pi Dt)^{-\frac{1}{2}} (e^{-\frac{\varepsilon^2}{4Dt}}(\frac{\varepsilon^2}{4Dt^2})) \\
    =(4\pi Dt)^{-\frac{1}{2}} e^{-\frac{\varepsilon^2}{4Dt}} (-\frac{1}{2t} + \frac{\varepsilon^2}{4Dt^2}) \\
    \frac{\partial G}{\partial t} = G(\varepsilon, t)(-\frac{1}{2t} + \frac{\varepsilon^2}{4Dt^2})
\end{align*}

Er Identisk så oppfyller G diffusjonsligningen. 

Viser så at $\phi$ løser diffusjonsligningen:

\begin{align*}
    \phi(x,t) &= \int_{-\infty}^{infty} h(y) G(x-y,t)dy \\
    \frac{\partial^2 \phi}{\partial x^2} &= \int h(y) \frac{\partial^2}{\partial x^2} G(x-y,t)dy
\end{align*}

Når vi deriverer $G(x-y,t)$ er det samme som å derivere $G(\xi, t)$, $(\xi = x - y)$

\begin{align*}
    \text{Altså:} \frac{\partial^2}{\partial x^2} G(x-y,t)=\frac{\partial^2}{\partial \xi^2} G(\xi, t) \\
    \text{Vi vet G oppfyller }\partial_t G = D \partial_{\xi \xi} G \text{ får vi: } \\
    \frac{\partial \phi}{\partial t} = \int h(y) D\frac{\partial^2}{\partial x^2} G(x-y,t)dy=D\frac{\partial^2 \phi}{\partial x^2} \\
    \Rightarrow \text{Dermed er }\phi \text{ en løsning av diffusjonsligningen.}
\end{align*}


Fysisk tolkning av h(y): Når $t \rightarrow 0^+$ blir $G(x-y,t)$ en veldig smal normalfordeling rundt $x=y$ (Dirachs deltafunksjon). Dermed gjelder

\begin{align*}
    G(\cdot , t) \rightarrow \delta(\cdot).
\end{align*}

Da får vi

\begin{align*}
    \phi (x,t) = \int_{-\infty}^{\infty} h(y) G(x-y, t)dy \xrightarrow[t \to 0^{+}] \int_{-\infty}^{infty} h(y) \delta (x-y) dy = h(x).
\end{align*}

Her er h(x) initialfordelingen av partikler ved starttid, det vil si

\begin{align*}
    \phi (x,0) = h(x)
\end{align*}

Altså beskriver h(y) hvor sannsynlig det er å finne en partikkel ved posisjon y når t = 0.

##### Oppgave 1b:

En partikkel starter på $x_0$ og med et steg, så kan den ende opp på $x$ $\epsilon$ \{$x_0 - 1$, $x_0$ , $x_0 + 1$\}.


Anta at for $P: x_0 \to x$, så er sannsynligheten for å finne en partikkel x i V(x) $P(x)=\frac{e^{-\beta V(x)}}{Z}$.



Vi antar også at det finnes en proporsjonal constant som er lik for alle steg muligheter.

La

\begin{align*}
    p^- =CP(x_0 -1), p^0 =CP(x_0), og p^+ =CP(x_0 +1)
\end{align*}

Her er $p^- + p^0 + p^+ =1$, og dette kan brukes for å løse for C. 

\begin{align*}
    1=C[P(x_0-1)+P(x_0)+P(x_0+1)]
\end{align*}

\begin{align*}
    C=\frac{1}{P(x_0-1)+P(x_0)+P(x_0+1)}
\end{align*}


Nå setter vi inn $\{x_0-1, x_0, x_0+1\}$ i P(x).

\begin{align*}
    P(x_0-1)=\frac{e^{-\beta V(x_0-1)}}{Z}  \\
    P(x_0)=\frac{e^{-\beta V(x_0)}}{Z}  \\
    P(x_0+1)=\frac{e^{-\beta V(x_0+1)}}{Z}
\end{align*}

Alle har felles nevner, så,

\begin{equation*}
    C=\frac{Z}{e^{-\beta V(x_0-1)}+e^{-\beta V(x_0)}+e^{-\beta V(x_0+1)}}
\end{equation*}


Nå kan vi sette konstanten vår C inn i $\{p^-,p^0,p^+\}$:

$p^-$:

\begin{align*}
    p^-=CP(x_0-1)=\frac{Z}{\Sigma e^{-\beta V}}\cdot\frac{e^{-\beta V(x_0-1)}}{Z}
\end{align*}

Z kansellerer, så vi får at

\begin{equation*}
    p^-=\frac{e^{-\beta V(x_0-1)}}{e^{-\beta V(x_0-1)}+e^{-\beta V(x_0)}+e^{-\beta V(x_0+1)}}
\end{equation*}

Deler både telleren og nevneren med $e^{-\beta V(x_0-1)}$


\begin{align*}
    \frac{e^{-\beta V(x_0-1)}}{e^{-\beta V(x_0-1)}}&=1   \\
    \frac{e^{-\beta V(x_0)}}{e^{-\beta V(x_0-1)}}&=e^{-\beta [V(x_0)-V(x_0-1)]}  \\
    \frac{e^{-\beta V(x_0+1)}}{e^{-\beta V(x_0-1)}}&=e^{-\beta [V(x_0+1)-V(x_0-1)]}
\end{align*}


Da får vi at 

\begin{equation*}
    p^-=\frac{1}{1+e^{-\beta [V(x_0)-V(x_0-1)]}+e^{-\beta [V(x_0+1)-V(x_0-1)]}}
\end{equation*}

Ved å gjøre dette med $p^0$ og $p^+$ så får vi at 

\begin{equation*}
    p^0=\frac{1}{1+e^{-\beta [V(x_0-1)-V(x_0)]}+e^{-\beta [V(x_0+1)-V(x_0)]}}
\end{equation*}

og

\begin{equation*}
    p^+=\frac{1}{1+e^{-\beta [V(x_0-1)-V(x_0+1)]}+e^{-\beta [V(x_0)-V(x_0+1)]}}
\end{equation*}


Dette var det vi skulle vise. $\square$



##### Oppgave 1c:

Skal finne ut av hva som skjer med sannsynlighetene $\{p^+,p^0,p^-\}$ fra oppgave 1b når temperaturen T oppfyller de følgende betingelsene: 

\begin{align*}
    \bullet \qquad& k_BT>>|V(x+1)-V(x)| \text{ for noen x}, \\
    \bullet \qquad& k_BT<<|V(x+1)-V(x) \text{ for noen x}.
\end{align*}

Boltzmanns vekt/faktor er definert som $e^{-\beta E_i}=e^{\frac{-E_i}{k_BT}}$. Dette ligner på sannsynligheten vi fikk i ligning 7 i oppgave 1b. 

Så, vi kan skrive om resultatene fra 1B om til Boltzmanns vekt form. 

La $w^+=e^{-\beta V(x_0+1)}$, $w^0=e^{-\beta V(x_0)}$, $w^-=e^{-\beta V(x_0-1)}$. 

Vi får da at $p^+=\frac{w^+}{w^-+w^0+w^+}$, $p^0=\frac{w^0}{w^-+w^0+w^+}$, og $p^-=\frac{w^-}{w^-+w^0+w^+}$.


##### $k_BT>>|V(x+1)-V(x)|$ for noen x:

Hvis T oppfyller $k_BT>>|V(x+1)-V(x)|$, så betyr det at vi ser på en høyere temperatur limit. 

Når temperaturen er høy, så er energiforskjellene mellom nabo punkter små når de er sammenlignet mde den termiske energien. Da er

\begin{equation*}
    e^{-\beta [V(x^i)-V(x^j)]}\approx 1
\end{equation*} 

for alle naboposisjoner. 

Her er, $w^- \approx w^0 \approx w^+$, så vi får at

\begin{align*}
    p^+ \approx p^0 \approx p^- \approx \frac{1}{3}
\end{align*}

Her kan vi se at partikkelen utfører tilnærmet en symmetrisk tilfeldig vandring, så potensialet har liten betydning for bevegelsen. 


##### $k_BT<<|V(x+1)-V(x)|$ for noen x

For lave temperaturer, så blir $\beta$ stor. Dette betyr at små energiforskjeller gir store forskjeller i Boltzmann faktorene. Da vil posisjonen med lavest potensialenergi dominere normaliseringen. 

Sannsynligheten for å bevege seg til posisjon med lavest $V(x)$ blant $\{x_0-1,x_0,x_0+1\}$ går mot 1, mens de andre går mot 0. 

Dette betyr at bevegelsen blir tilnærmet deterministisk, og partikkelen beveger seg nedover i potensiallandskapet. 


Så, i konklusjon så ser vi at

\begin{align*}
    \bullet \qquad &\text{ For høy temperatur, så kan potensialet neglisjeres, og problemet forenkles til vanlig diffusjon} \\
    &\text{ med konstante overgangssansynligheter.} \\
    \bullet \qquad &\text{ For lav temperatur, så kan den termiske støyen neglisjeres, og bevegelsen kan forenkles til at} \\
    &\text{ partikkelen alltid beveger seg mot lavere potensialenergi. }
\end{align*}

```python
### Exercise 2a)

N_particles = 101
N_steps = 200
h = 1
beta_k = 1
N_runs = 100

initial_positions = np.arange(N_particles)

#skal starte med i) V(x)=k:

def V_constant(x,k=1):
    return k

def transition_probabilities(x0, V, beta=1.0):
    V_minus = V(x0 - 1)
    V_0 = V(x0)
    V_plus = V(x0 + 1)

    w_minus = np.exp(-beta * V_minus)
    w_0 = np.exp(-beta * V_0)
    w_plus = np.exp(-beta * V_plus)

    Z = w_minus + w_0 + w_plus

    p_minus = w_minus / Z
    p_0 = w_0 / Z
    p_plus = w_plus / Z

    return p_minus, p_0, p_plus

def step(positions, V, beta=1.0):
    new_positions = positions.copy()
    for i, x0 in enumerate(positions):
        p_minus, p_0, p_plus = transition_probabilities(x0, V, beta)
        r = np.random.rand()
        if r <= p_minus:
            new_positions[i] = x0 - h
        elif r > 1 - p_plus:
            new_positions[i] = x0 + h
        else:
            new_positions[i] = x0
    return new_positions

counts_sum = {}
min_x = None
max_x = None

for _ in range(N_runs):
    positions = initial_positions.copy()

    for _ in range(N_steps):
        positions = step(positions, V_constant, beta=beta_k)

    xs, counts = np.unique(positions, return_counts=True)

    if min_x is None:
        min_x, max_x = int(xs.min()), int(xs.max())
    else:
        min_x = min(min_x, int(xs.min()))
        max_x = max(max_x, int(xs.max()))

    for x, c in zip(xs, counts):
        counts_sum[int(x)] = counts_sum.get(int(x), 0) + int(c)
```

```python
#plot for 2a(i):

x_axis = np.arange(min_x, max_x + 1)
avg_counts = np.array([counts_sum.get(int(x), 0) for x in x_axis], dtype=float) / N_runs
avg_density = avg_counts / N_particles


plt.figure()
plt.plot(x_axis, avg_density, marker='o', linestyle='-')
plt.xlabel("x")
plt.ylabel("Gjennomsnittlig partikkelfordeling (sannsynligheten)")
plt.title("Oppgave 2a(i): $V(x) = k$")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
###Del ii) V(x)=-kx:


def V_linear(x, k=1):
    return -k * x

counts_sum_linear = {}
min_x_linear = None
max_x_linear = None

for _ in range(N_runs):
    positions = initial_positions.copy()

    for _ in range(N_steps):
        positions = step(positions, V_linear, beta=beta_k)
    
    xs, counts = np.unique(positions, return_counts=True)

    if min_x_linear is None:
        min_x_linear, max_x_linear = int(xs.min()), int(xs.max())
    else:
        min_x_linear = min(min_x_linear, int(xs.min()))
        max_x_linear = max(max_x_linear, int(xs.max()))

    for x, c in zip(xs, counts):
        counts_sum_linear[int(x)] = counts_sum_linear.get(int(x), 0) + int(c)


x_axis_linear = np.arange(min_x_linear, max_x_linear + 1)
avg_counts_linear = np.array(
    [counts_sum_linear.get(int(x), 0) for x in x_axis_linear], dtype=float
) / N_runs
avg_density_linear = avg_counts_linear / N_particles

plt.figure()
plt.plot(x_axis_linear, avg_density_linear, marker='o', linestyle='-')
plt.xlabel("x")
plt.ylabel("Gjennomsnittlig partikkelfordeling (sannsynlighet)")
plt.title("Oppgave 2a(ii): $V(x) = -kx$")
plt.grid(True, alpha=0.3)
plt.show()

```

##### Oppgave 2b:

Her simulerer vi et tidssteg av en 1D random walk for mange partikler, men med hard core repulsion. Altså ingen to partikler får stå på samme posisjon. I hvert tidssteg flytter alle partikler seg én gang. Rekkefølgen er tilfeldig (for å unngå bias). Hver partikkel velger venstre/stå/høyre basert på sannsynligheter fra potensialet V(x). Dersom målcellen er opptatt vil flyttet avbrytes. For å sjekke om en posisjon allerede er opptatt bruker vi et "set" med okkuperte posisjoner, siden oppslag i et "set" er raskere enn i en liste. Occupied er mye raskere enn candidate in new\_positions som vi brukte først.

```python
### Exercise 2b

def step_hardcore(positions, V, beta=1.0, h=1):
    new_positions = positions.copy()
    occupied = set(new_positions.tolist())

    indices = np.random.permutation(len(new_positions))

    for i in indices:
        x0 = new_positions[i]

        p_minus, p_0, p_plus = transition_probabilities(x0, V, beta)
        r = np.random.rand()

        if r <= p_minus:
            candidate = x0 - h
        elif r > 1 - p_plus:
            candidate = x0 + h
        else:
            candidate = x0

        if candidate != x0 and candidate not in occupied:
            occupied.remove(x0)
            occupied.add(candidate)
            new_positions[i] = candidate

    return new_positions


def run_simulation_avg_density(step_fn, V_fn, beta=1.0):
    counts_sum = {}
    min_x = None
    max_x = None

    for _ in range(N_runs):
        positions = initial_positions.copy()

        for _ in range(N_steps):
            positions = step_fn(positions, V_fn, beta=beta, h=h)

        xs, counts = np.unique(positions, return_counts=True)

        if min_x is None:
            min_x, max_x = int(xs.min()), int(xs.max())
        else:
            min_x = min(min_x, int(xs.min()))
            max_x = max(max_x, int(xs.max()))

        for x, c in zip(xs, counts):
            counts_sum[int(x)] = counts_sum.get(int(x), 0) + int(c)

    x_axis = np.arange(min_x, max_x + 1)
    avg_counts = np.array([counts_sum.get(int(x), 0) for x in x_axis], dtype=float) / N_runs
    avg_density = avg_counts / N_particles

    return x_axis, avg_density
```

```python
# Graf for 2b(i) og for V(x) = k 
x_axis_b1, avg_density_b1 = run_simulation_avg_density(
    step_fn=step_hardcore,
    V_fn=V_constant,
    beta=beta_k
)

plt.figure()
plt.plot(x_axis_b1, avg_density_b1, marker='o', linestyle='-')
plt.xlabel("x")
plt.ylabel("Gjennomsnittlig partikkelfordeling (sannsynlighet)")
plt.title("Oppgave 2b(i): Hard-core, $V(x)=k$")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Graf for 2b(ii)for V(x) = -kx

x_axis_b2, avg_density_b2 = run_simulation_avg_density(
    step_fn=step_hardcore,
    V_fn=V_linear,
    beta=beta_k
)

plt.figure()
plt.plot(x_axis_b2, avg_density_b2, marker='o', linestyle='-')
plt.xlabel("x")
plt.ylabel("Gjennomsnittlig partikkelfordeling (sannsynlighet)")
plt.title("Oppgave 2b(ii): Hard-core, $V(x)=-kx$")
plt.grid(True, alpha=0.3)
plt.show()
```

Her ser vi at for $V(x) = k$ er potensialet konstant, og partiklene har derfor ingen preferanse for å bevege seg i en bestemt retning. Fordelingen blir derfor symmetrisk rundt startområdet. Sammenlignet med tilfellet uten vekselvirkning fører hard-core repulsjon til at partiklene ikke kan samle seg på samme startposisjon, noe som gjør fordelingen noe bredere og flatere.

##### Oppgave 2c:

Når temperaturen senkes slik at ($\beta k = 1000$), vil vi få svært lav temperatur (stor $\beta$). Da blir overgangssannsynligheten 


\begin{align*}
    p \propto e^{-\beta V(x)}
\end{align*}

og vil bli svært følsomme for forskjeller i potensialet. Dette betyr at partiklene nesten alltid vil bevege seg i retningen som gir lavest potensialenergi.


For potensialet $V(x)=k$ er potensialet konstant for alle posisjoner. Dermed er

\begin{align*}
    V(x-1)=V(x)=V(x+1),
\end{align*}

og sannsynlighetene for å gå til venstre, bli stående eller gå til høyre blir fortsatt

\begin{align*}
    p^- = p^0 = p^+ = \frac{1}{3}.
\end{align*}


En lavere temperatur endrer derfor ikke bevegelsen i dette tilfellet. Partiklene utfører fortsatt en symmetrisk tilfeldig vandring. I tilfelle 2b vil hard-core-interaksjonen fortsatt hindre at flere partikler kan være på samme posisjon, noe som gjør fordelingen litt mer spredt enn i tilfellet uten interaksjon.

For potensialet $V(x)=-kx$ blir potensialet lavere når $x$ øker. Når $\beta k$ er veldig stor, vil sannsynligheten for å gå mot høyre bli mye større enn de andre. I praksis betyr dette at

\begin{align*}
    p^+ \approx 1, \qquad p^- \approx 0, \qquad p^0 \approx 0.
\end{align*}

Partiklene vil derfor nesten alltid bevege seg mot høyre ved hvert steg. I tilfellet uten interaksjon (2a) vil hele partikkelfordelingen derfor flytte seg mot større $x$. I tilfellet med hard-core-interaksjon (2b) kan partiklene ikke passere hverandre, og de bakerste partiklene kan bli blokkert av de som står foran. Resultatet er at partiklene beveger seg som en mer samlet gruppe som gradvis flytter seg mot høyre. Systemet blir mer deterministisk (mindre tilfeldig).


### Oppgave 3 - Random walk in a ratchet potential without interactions

```python
#Oppgave 3a

Nx = 100
L = 2 * Nx
Tp = 500
T_total = 20 * Tp
k = 1.0
beta_k = 1000
beta = beta_k / k
Np = 12 * Nx
rng = np.random.default_rng(0)



def V1_sawtooth(x, Nx=100, alpha=0.8, k=1):
    x = np.asarray(x)

    u = np.mod(x, Nx).astype(float)

    y = np.where(u <= alpha * Nx, u, u - Nx)

    V = np.zeros_like(x, dtype=float)

    mask_pos = y > 0
    V[mask_pos] = k * (y[mask_pos] / (alpha * Nx)) #fra ligning 9 i oppgaven.

    mask_neg = ~mask_pos
    V[mask_neg] = (-k) * (y[mask_neg] / ((1 - alpha) * Nx)) #igjen fra ligning 9.

    return V

def step_flat(pos):
    r = rng.random(size=pos.size)
    move = np.zeros_like(pos, dtype=int)
    move[r < 1/3] = -1
    move[r > 2/3] = +1
    n_minus = np.count_nonzero(move == -1)
    n_plus = np.count_nonzero(move == +1)
    pos = (pos + move) % L
    return pos, n_plus, n_minus

def step_saw(pos, alpha):
    xminus = (pos - 1) % L
    x0 = pos
    xplus = (pos + 1) % L

    Vminus = V1_sawtooth(xminus, Nx, alpha = alpha, k = k)
    V0 = V1_sawtooth(x0, Nx, alpha = alpha, k = k)
    Vplus = V1_sawtooth(xplus, Nx, alpha = alpha, k = k)

    Vmin = np.minimum(np.minimum(Vminus, V0), Vplus) #Prøver å unngå overflow ved å trekke fra min før eksponentiering. 

    wminus = np.exp(-beta * (Vminus - Vmin))
    w0 = np.exp(-beta * (V0 - Vmin))
    wplus = np.exp(-beta * (Vplus - Vmin))

    Z = wminus + w0 + wplus

    pminus = wminus / Z
    pplus = wplus / Z

    r = rng.random(size=pos.size)
    move = np.zeros_like(pos, dtype=int)
    move[r <= pminus] = -1
    move[r > (1 - pplus)] = +1

    nminus = np.count_nonzero(move == -1)
    nplus = np.count_nonzero(move == +1)
    pos = (pos + move) % L
    return pos, nplus, nminus


def simulate_a(alpha):
    pos = np.repeat(np.arange(L), Np // L)

    J_t = np.zeros(T_total, dtype=float)

    for t in range(T_total):
        in_V2 = ((t // Tp) % 2 == 0) #starter i V2 med t=0, og vi bytter for hver Tp.
        if in_V2:
            pos, nplus, nminus = step_flat(pos)
        else:
            pos, nplus, nminus = step_saw(pos, alpha)
        
        J_t[t] = (nplus - nminus) / Np

    J_avg = np.array([J_t[n*(2*Tp):(n+1)*(2*Tp)].mean() for n in range(10)])
    return J_t, J_avg


results = {}

for alpha in [0.8, 0.1]:
    J_t, J_avg = simulate_a(alpha)
    results[alpha] = (J_t, J_avg)

    print(f"\nalpha = {alpha}:")
    for i, val in enumerate(J_avg):
        print(f"  J_avg({i}) = {val:.6e}")

for alpha in [0.8, 0.1]:
    J_t, J_avg = results[alpha]
```

### Oppgave 4

Har parameterne er $\beta$k = 1000, $N_x$=20, $\alpha$ = 0.2 og $T_p$ = 40. Vi har en x-akse som rommer fire hele "sawtooth peaks", altså total lengde blir $L= 4N_x= 80$.
Vi kjører simuleringen i fem sykler, totalt $T= 5T_P = 200$ tiddssteg, og bruker periodiske ransbetingelser. 

Valg av partikler, størrelse og initialtilstand:

Vi velger N= 10 partikler og parikkelstørrelse b=2. Antallet partikler velges slik at systemet ikke blir for tettpakket, samtidig som interaksjoner mellom partiklene fortsatt spiller en rolle. Initialt plasseres partikler jevnt fordelt på ringen, f.eks. ved posisjonene {0,8,16,...,72} slik at avstanden mellom naboparikler er større enn b.

```python
### oppgave 4a


beta = 1000.0          
Nx = 20
alpha = 0.2
Tp = 40

n_peaks = 4
L = Nx * n_peaks       # 80 totalt
n_cycles = 5
T = n_cycles * Tp      # 200 steg

# Valgte verdier 
N = 10                 
b = 2                  

def sawtooth_V_at(x, Nx=20, alpha=0.2, k=1.0):
    """
    Sawtooth potential over one period Nx, extended periodically.
    Returns V(x) for integer x (site index).
    """
    xm = x % Nx
    m = int(round(alpha * Nx))
    m = max(1, min(Nx - 1, m))

    
    if xm < m:
        return k * (xm / m)                         
    else:
        return k * ((Nx - xm) / (Nx - m))          


def flashing_V_at(x, t, Tp=40, Nx=20, alpha=0.2, k=1.0):
    """
    Flashing: ON for half period, OFF for half period.
    """
    on = (t % Tp) < (Tp // 2)
    if on:
        return sawtooth_V_at(x, Nx=Nx, alpha=alpha, k=k)
    else:
        return 0.0



def ring_distance(a, c, L):
    d = abs(a - c)
    return min(d, L - d)


def allowed_position(candidate, positions, i, b, L):
    """
    True if particle i can be at 'candidate' without being within
    distance <= b of any other particle, on a ring of length L.
    """
    for j, xj in enumerate(positions):
        if j == i:
            continue
        if ring_distance(candidate, xj, L) <= b:
            return False
    return True


def transition_probabilities(x0, t, beta, L, Tp, Nx, alpha, k=1.0):
    xL = (x0 - 1) % L
    xR = (x0 + 1) % L

    V0 = flashing_V_at(x0, t, Tp=Tp, Nx=Nx, alpha=alpha, k=k)
    VL = flashing_V_at(xL, t, Tp=Tp, Nx=Nx, alpha=alpha, k=k)
    VR = flashing_V_at(xR, t, Tp=Tp, Nx=Nx, alpha=alpha, k=k)

    wL = np.exp(-beta * (VL - V0))
    wR = np.exp(-beta * (VR - V0))
    w0 = 1.0

    Z = wL + w0 + wR
    p_minus = wL / Z
    p_plus = wR / Z
    p_0 = w0 / Z
    return p_minus, p_0, p_plus


def step_hardcore_sizeb(positions, t, beta, b, L, Tp, Nx, alpha, k=1.0):
    positions = positions.copy()
    order = np.random.permutation(len(positions))

    for i in order:
        x0 = positions[i]
        p_minus, p_0, p_plus = transition_probabilities(
            x0, t, beta=beta, L=L, Tp=Tp, Nx=Nx, alpha=alpha, k=k
        )

        r = np.random.rand()
        if r <= p_minus:
            cand = (x0 - 1) % L
        elif r > 1.0 - p_plus:
            cand = (x0 + 1) % L
        else:
            cand = x0

        if cand != x0 and allowed_position(cand, positions, i, b=b, L=L):
            positions[i] = cand

    return positions


def simulate(positions0, T, beta, b, L, Tp, Nx, alpha, k=1.0):
    traj = np.zeros((T + 1, len(positions0)), dtype=int)
    traj[0] = positions0.copy()

    pos = positions0.copy()
    for t in range(T):
        pos = step_hardcore_sizeb(pos, t, beta, b, L, Tp, Nx, alpha, k=k)
        traj[t + 1] = pos

    return traj


# Run + plot 
def main():
    # Valgte initialer (b=2, L=80)
    positions0 = np.array([8 * i for i in range(N)], dtype=int)

    for i in range(N):
        for j in range(i + 1, N):
            if ring_distance(positions0[i], positions0[j], L) <= b:
                raise ValueError("Initial positions violate hard-core condition.")

    traj = simulate(positions0, T=T, beta=beta, b=b, L=L, Tp=Tp, Nx=Nx, alpha=alpha, k=1.0)

    plt.figure()
    for i in range(traj.shape[1]):
        plt.plot(traj[:, i], lw=1)
    plt.xlabel("time step")
    plt.ylabel("position (site index)")
    plt.title(f"Exercise 4a: flashing ratchet, beta*k=1000, b={b}, N={N}, L={L}")
    plt.show()
    
if __name__ == "__main__":
    main()
```

Hvordan partiklene beveger seg i det potensielle landskap:

Hvis vi ser på figuren vi får ut, forteller den oss hvor hver partikkel er over tid. Hver farge er én partikkel. Når vi ser brå hopp opp eller ned, betyr det at partikkelen flytter seg til en ny binding site. Når ratchet-potensialet er on, blir partiklene ofte fanget i lokale minima, og posisjoene endrer seg lite. Når potensialet er off, kan partiklene diffudere friere, og vi ser større hopp i posisjon mellom binding sites. På grunn av asymmetrien i sawtooth potensial oppstår det en netto drift i én retning over flere sykluser. Hard core repulsion gjør at partiklene ikke kan overlappe eller passere hverandre, slik at rekkefølge mellom partiklene langs ringen i stor grad bevares.

#### oppgave 4b:


Når vi har Nc= 100 sykler for hver tetthet p tar koden ganske lang tid. Når tettheten er lav er det få partikler i systemet, de er langt fra hverandre og Hard-core repulsion spiller ingen rolle. Fordi det er så få partikler, blir målingene av current veldig støyete. Vil da få store error og litt tilfeldig verdi av $J$. 

Ved middels tetthet (toppen av grafen) vil flere parikler delta i systemet, statistikken blir bedre og ratchet-mekanismen virker mer stabilt. Her skjer det viktigste i modellen. Når potensialet er "on", sitter partiklene i lokale minima. Når potensialet slås "off", kan de diffudere. Når potensialet slås "on" igjen, faller de ned i nye minima. Vil da få positiv current og grafen får en topp rundt $p= 0.5$ ca. 

Når tettheten blir høy vil partiklene ligge veldig tett, hard-core repulsion hindrer dem i å passere hverandre og forsøk på å flytte blir avbrutt. Systemet blir fullt, og partiklene har ingen plass å flytte seg. Ser at $J$ går mot null når $\rho$ går mot 1. Det er det grafen viser

```python
#oppgave 4b

beta = 1000.0
Nx = 100
Tp = 300
Ns = 10
alpha = 0.2
b = 20

L = Ns * Nx         
Nc = 100
T = Nc * Tp         #totalt steg

def sawtooth_V_at(x, Nx=100, alpha=0.2, k=1.0):
    xm = x % Nx
    m = int(round(alpha * Nx))
    m = max(1, min(Nx - 1, m))
    if xm < m:
        return k * (xm / m)
    else:
        return k * ((Nx - xm) / (Nx - m))
    
# sawtooth når den er på, for x=0,...,L-1
V_one_period = np.array([sawtooth_V_at(x, Nx=Nx, alpha=alpha, k=1.0) for x in range(Nx)])
V_on = np.tile(V_one_period, Ns)   # lengde L

def ring_distance(a, c, L):
    d = abs(a - c)
    return min(d, L - d)

def allowed_position(candidate, positions, i, b, L):
    if len(positions) <= 1:
        return True

    others = np.delete(positions, i)
    d = np.abs(others - candidate)
    d = np.minimum(d, L - d)
    return np.all(d >= b) #skal være raskere enn en for løkke


def transition_probabilities(x0, t, beta, L, Tp, V_on):
    xL = (x0 - 1) % L
    xR = (x0 + 1) % L

    on = (t % Tp) < (Tp // 2)
    if on:
        V0 = V_on[x0]
        VL = V_on[xL]
        VR = V_on[xR]
    else:
        V0 = VL = VR = 0.0

    wL = np.exp(-beta * (VL - V0))
    wR = np.exp(-beta * (VR - V0))
    w0 = 1.0
    Z = wL + w0 + wR
    
    return wL / Z, w0 / Z, wR / Z


def step_hardcore_sizeb_and_disp(positions, t, beta, b, L, Tp, V_on):
    positions = positions.copy()
    order = np.random.permutation(len(positions))

    disp_sum = 0  

    for i in order:
        x0 = positions[i]
        p_minus, p_0, p_plus = transition_probabilities(x0, t, beta, L, Tp, V_on)

        r = np.random.rand()
        if r <= p_minus:
            cand = (x0 - 1) % L
        elif r > 1.0 - p_plus:
            cand = (x0 + 1) % L
        else:
            cand = x0

        if cand != x0 and allowed_position(cand, positions, i, b=b, L=L):
            if cand == (x0 + 1) % L:
                disp_sum += 1
            elif cand == (x0 - 1) % L:
                disp_sum -= 1
            positions[i] = cand

    return positions, disp_sum


def run_current_for_density(Np, beta, b, L, Tp, Nc, V_on, seed=None):
    if seed is not None:
        np.random.seed(seed)

    positions = np.floor(np.linspace(0, L, Np, endpoint=False)).astype(int)
   
    for i in range(Np):
        for j in range(i + 1, Np):
            if ring_distance(positions[i], positions[j], L) < b:
                raise ValueError("Initial positions violate hard-core (distance < b).")

    J_cycles = []
    t_global = 0

    for _ in range(Nc):
        disp_cycle = 0
        
        for _ in range(Tp):
            positions, disp_step = step_hardcore_sizeb_and_disp(
                positions, t_global, beta, b, L, Tp, V_on
            )
            disp_cycle += disp_step
            t_global += 1

        
        J_cycles.append(disp_cycle / (Np * Tp))

    return float(np.mean(J_cycles)), float(np.std(J_cycles))

def main_4b():
    Np_max = L // b  
    Np_values = np.unique(np.linspace(1, Np_max, 12, dtype=int))

    rhos = (b * Np_values) / L
    Js = []
    Jerrs = []

   
    for Np, rho in zip(Np_values, rhos):
        J_mean, J_std = run_current_for_density(
            Np, beta=beta, b=b, L=L, Tp=Tp, Nc=Nc, V_on=V_on
        )
        Js.append(J_mean)
        Jerrs.append(J_std)

        print(f"Np={Np:2d}, rho={rho:.2f}, J={J_mean:.5f} ± {J_std:.5f}")

    Js = np.array(Js)
    Jerrs = np.array(Jerrs)

    plt.figure()
    plt.errorbar(rhos, Js, yerr=Jerrs, fmt='o-', capsize=3)
    plt.xlabel("density $\\rho$")
    plt.ylabel("cycle-averaged current $J$")
    plt.title("Exercise 4b: cycle-averaged current vs density")
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    main_4b()
```

Men siden OFF-perioden er så kort, rekker ikke partiklene å flytte seg noe langt. Når potensialet slår på igjen, sitter de nesten på samme sted. Ratchet-effekten blir svak, derfor liten current.

Et eksempel vi fant på forskning som studerer partikler med interaksjoner i ratchet-systemer er "Review-artikkelen Ratchet Effects in Active Matter Systems" av C.J. Olson Reichhardt og C.Reichhardt (2017). Artikkelen beskriver hvordan asymmetriske barrierer kan føre til rettet transport av partikler når systemet drives ut av likevekt. 

I artikkelen vises et eksperiment der passive kolloidpartikler befinner seg i et bad av svømmende bakterier i en geometri med asymmetriske ratchet-barrierer (figur 12). Figuren viser hvordan antallet partikler i ulike kamre i systemet utvikler seg over tid. På grunn av den asymmetriske geometrien kan partiklene lettere passere barrierene i én retning enn i den andre. Dette fører til en netto transport av partikler og en opphopning i bestemte områder i systemet. 

Modellen består av passive kolloidpartikler som påvirkes av svømmende bakterier. Bakteriene driver systemet bort fra likevekt og gir partiklene en effektiv bevegelse gjennom ratchet-strukturen.

Transporten skyldes samspillet mellom passive partikler og de aktive bakteriene, som driver systemet bort fra likevekt. Interaksjoner mellom partiklene spiller derfor en viktig rolle for hvordan partiklene fordeler seg og hvordan transporten oppstår. Uten slike iteraksjoner ville partiklene bevege seg mer uavhengig av hverandre, og effekter som blokkering og opphopning ville vært svakere.

Dette kan sammenlignes med simuleringen i denne oppgaven, der partikler beveger seg i et flashing ratchet-potensial og interagerer gjennom hard-core repulsjon. Ved lave tettheter kan partiklene bevege seg relativt fritt, mens ved høyere tetthet vil mange hopp bli blokkert av andre partikler. Dette fører til redusert strøm ved høy tetthet, noe som er et eksempel på hvordan interaksjoner mellom partikler kan påvirke ratchet-transporten.
